```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
input_file = '한국_기업문화_HR_데이터셋_샘플.csv'
df = pd.read_csv(input_file)
df.head()
```

```python
print(df.columns)
```

```python
#데이터 전처리 (15점)
print(df.isna().sum())
# Nan 값이 없음 -> 결측치 처리 안함

#고유값 확인
for col in df:
    print(df[col].unique())

df_change_num = df[['Age', '집까지거리', '학력수준', '근무환경만족도', '시간당급여', '직급', '업무만족도', '월급여']]
df_change_cat = df[['출장빈도', '야근여부', '이직여부']]
# 직원들의 이직과 관련된 내용들, 대부분 직원들의 개인 정보와 관련된 10가지 콜럼을 선택해서 새로운 데이터프레임을 만듦


# 이직여부를 이진값(0, 1)으로 변환하시오.
le = LabelEncoder()
for col in df_change_cat:
    df_change_cat[col] = le.fit_transform(df_change_cat[col])
df_change_cat
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
df_num_mms = mms.fit_transform(df_change_num)
df_num_mms_pd = pd.DataFrame(df_num_mms, columns=df_change_num.columns)
df_num_mms_pd.head()
```

```python
df_processed = pd.concat([df_num_mms_pd, df_change_cat], axis=1)
df_processed.head()
print(len(df_processed))
```

```python
raw = df_processed
np_raw = raw.values
type(np_raw)
```

```python
X = df_processed.drop('이직여부', axis=1)
y = df_processed['이직여부']
```

```python
train = np_raw[:800]
test = np_raw[800:]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
y_train = [i[0] for i in train]
X_train = [j[1:] for j in train]

y_test = [i[0] for i in test]
X_test = [j[1:] for j in test]

len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion='entropy', max_depth=3, min_samples_leaf=5)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Test Accuracy is: ", accuracy_score(y_test, y_pred)*100)
```
